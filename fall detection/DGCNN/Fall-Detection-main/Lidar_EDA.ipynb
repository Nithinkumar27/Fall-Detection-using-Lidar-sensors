
# Exploratory Data Analysis (EDA) for LiDAR Datasets
    


## 1. Data Loading and Initial Inspection
    

```python

import pandas as pd

# Load the dataset
file_path = 'raw_walking (Frame 26).csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
    
```

## 2. Data Cleaning
    

```python

# Check for missing values
missing_values = data.isnull().sum()
missing_values
    
```

## 3. Univariate Analysis
    

```python

# Example: Distribution of Intensity
data['intensity'].hist(bins=50)
    
```

## 4. Bivariate and Multivariate Analysis
    

```python

# Example: Scatter plot for Intensity vs Distance
data.plot.scatter(x='distance_m', y='intensity', alpha=0.5)
    
```

## 5. Visualization
    

```python
import plotly.express as px

fig = px.scatter_3d(data, 
                    x='Points_m_XYZ:0', 
                    y='Points_m_XYZ:1', 
                    z='Points_m_XYZ:2', 
                    color='intensity', 
                    size_max=10, 
                    opacity=0.7)

fig.update_layout(title="3D Visualization of LiDAR Point Cloud")
fig.show()
```
